# Working with images in Python

```python
from skimage import data 
import numpy as np

cat = data.chelsea()
print('The size of the image is {}'.format(np.shape(cat)))
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 

plt.imshow(cat)
```

```python
redComp = cat[:,:,0]
plt.imshow(redComp)
```

```python
plt.imshow(redComp, cmap='ocean')
plt.colorbar()
```

```python
import matplotlib.image as img

prism = img.imread('color.jpg')
plt.imshow(prism)

plt.colorbar()
```

```python
redComp = prism[:,:,0]
plt.imshow(redComp, cmap='binary')
plt.colorbar()
```

```python
blueComp = prism[:,:,1]
plt.imshow(blueComp, cmap='binary')
plt.colorbar()
```

```python
greenComp = prism[:,:,2]
plt.imshow(greenComp, cmap='binary')
plt.colorbar()
```

```python
odd_row = [0.6,0]*5
even_row = [0,0.6]*5
chess= [odd_row,even_row]*5
print(chess)
plt.imshow(chess, cmap='binary')
plt.colorbar()
```

```python
total = [np.add(chess,0), np.add(chess, 0), np.add(chess, 0.3)]
print(np.shape(total))
total = np.transpose(total)
print(np.shape(total))

plt.imshow(total)
```

# water image

```python
water = img.imread('water.jpg')
plt.imshow(water)
```

```python
# histogram
n,b,p=plt.hist(water.ravel(), bins=100, range=(0,100), normed=True)
```
